```python
gdown.download("https://drive.google.com/uc?id=1SkCbZdnzng1a43m_vRRrVbmeFOPcyIU0", "cancer.csv", quiet=False)
cancer = pd.read_csv('cancer.csv')
```

```python
cancer.head()
cancer = cancer.set_index("State", drop=True)
display(cancer)
```

```python
cancer.isnull()
cancer["Rate"] = cancer["Rate"].fillna(cancer["Rate"].mean())
display(cancer)
```

```python
def iqr_outlier(data):
    for i in range(len(data.iloc[1])):
        cancer_column = data.iloc[:, i]
        if cancer_column.apply(lambda x: isinstance(x, (int, float))).all():
            
            q1, q3 = np.percentile(data.iloc[:, i], [25, 75])
            print(f"q1: {q1}")
            print(f"q3: {q3}")

            iqr = q3 - q1
            lower = q1 - (iqr * 1.5)
            upper = q3 + (iqr * 1.5)
            print(f"lower: {lower}")
            print(f"upper: {upper}")

            data = data[data.iloc[:, i] < upper]
            data = data[data.iloc[:, i] > lower]
        
        data = data.reset_index(drop=True)
        return data

iqr_outlier(cancer)
```
