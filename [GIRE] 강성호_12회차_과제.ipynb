## NAND Gate

NAND Gate란, AND gate의 역으로써 x_1, x_2가 모두 1일 때만 0을 반환하고, 나머지는 1을 반환하는 논리 회로이다.  
아래 빈칸을 채워 NAND gate를 구현하시오.

```python
def NAND_gate(x1, x2):
    w1 = -0.3
    w2 = -0.3
    b = 0.2

    if x1 * w1 + x2 * w2 >= -b:
        return 1
    else:
        return 0

for x1 in (0, 1):
    for x2 in (0, 1):
        print(f"NAND({x1}, {x2}): {NAND_gate(x1, x2)}")
```

이번에는 Keras를 활용해 구현하시오.

```python
NAND_gate_model = keras.models.Sequential([
    keras.layers.InputLayer(
        shape=(2,),
    ),
    keras.layers.Dense(
        units=1,
        activation=keras.activations.sigmoid
    )
])

NAND_gate_model.summary()
```

```python
import numpy as np

prob = NAND_gate_model(np.array([[0, 0], [0, 1], [1, 0], [1, 1]])).numpy()


display(prob)

def prob_to_pred(prob):
    if prob >= 0.5:
        return 1
    else:
        return 0


display(np.apply_along_axis(prob_to_pred, 1, prob))

NAND_gate_model.compile(optimizer='adamw', loss=keras.losses.BinaryCrossentropy, metrics=['accuracy'])

NAND_gate_model.fit(np.array([[0, 0], [0, 1], [1, 0], [1, 1]]), np.array([[1], [0], [0], [0]]), epochs=700)
```

```python
prob_after_training = NAND_gate_model(np.array([[0, 0], [0, 1], [1, 0], [1, 1]])).numpy()

display(prob_after_training)
display(np.apply_along_axis(prob_to_pred, 1, prob_after_training))
display(NAND_gate_model.weights)
```
